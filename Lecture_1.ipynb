But in the quantum world---the world that emerges down near the scale of atoms---things are different. Not spooky or incomprehensible. Just way different.
### Uncertainity Principle 
- Firstly in quantum mechanics you cannot calculate exact trajectory of a particle (atoms, electrons etc). What you can calculate is the probability of the presence of particle
- There are some complimentary quantities which follow the Heisenberg's uncertainity principle - For example, for a particle following "Quantum Rules", you cannot find the position and the momentum of the particle simultaneosuly.
- For example Shining a Light to "see" a tiny particle like an electron, we shine light on it.
    - To see it more clearly, we use light with shorter wavelength (more precise).
    - But shorter wavelength = higher energy photons.
    - These photons kick the particle, changing its speed.
So, the better we try to see where it is, the more we disturb its motion.

This is not a flaw in the microscope — it’s how quantum nature behaves.

This innate uncertainty---and the accompanying probabilities---are core features of quantum physics. But there are other things that set the quantum world apart from the classical world.
#### Mathematical Form
$
\Delta x \cdot \Delta p \geq \frac{\hbar}{2}
$
#### What it means:
- $\Delta x$: The **uncertainty in position** (how precisely we know where the particle is).
- $\Delta p$: The **uncertainty in momentum** (how precisely we know how fast and in which direction it's moving).
- $\hbar$: Called "h-bar", it’s a very small constant:

  $
  \hbar = \frac{h}{2\pi} \approx 1.055 \times 10^{-34} \ \text{J·s}
  $

- The product of the two uncertainties can **never be smaller** than $\frac{\hbar}{2}$.

### Quantization 
- Imagine you have a speaker with a volume knob. In most real life systems, you can turn the volume smoothly — just a little louder or softer. That’s called a continuous change.
- Instead of a smooth knob, the volume only changes in steps — level 1, level 2, level 3 — and nothing in between.

That’s quantisation:
A system where certain quantities can only take specific, discrete values — no in-betweens.

💡 Everyday Analogy: Stairs vs Ramp
A ramp: you can stand anywhere — smooth and continuous.

Stairs: you can only stand on specific steps — these are quantised levels.

In classical physics, energy is like a ramp: it can change in tiny continuous amounts.

In quantum physics, energy comes in steps — just like stairs.

## Interactive plot of a classical superposition of two waves

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output

# Parameters
x = np.linspace(0, 4 * np.pi, 500)
A1 = 1.0
A2 = 1.0
k = 2 * np.pi / 4
omega = 2 * np.pi / 30
phase_difference = np.pi # try 0 or np.pi too

# Loop over time steps and redraw plot
for t in range(100):
    y1 = A1 * np.sin(k * x - omega * t)
    y2 = A2 * np.sin(k * x - omega * t + phase_difference)
    y_total = y1 + y2

    clear_output(wait=True)
    plt.figure(figsize=(8, 4))
    plt.plot(x, y1, '--', label='Wave 1')
    plt.plot(x, y2, '--', label='Wave 2')
    plt.plot(x, y_total, 'k-', linewidth=2, label='Superposition')
    plt.ylim(-2.5, 2.5)
    plt.xlim(x[0], x[-1])
    plt.title('Superposition of Two Waves')
    plt.xlabel('x')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend(loc='upper right')
    plt.show()
    time.sleep(0.05)
```

# Quick Mathematical Derivations

Consider two sinusoidal waves:

$$
y_1(x, t) = A \cdot \sin(kx - \omega t)
$$

$$
y_2(x, t) = A \cdot \sin(kx - \omega t + \phi)
$$

By the principle of superposition:

$$
y_{\text{total}}(x, t) = y_1 + y_2 = A \cdot \sin(kx - \omega t) + A \cdot \sin(kx - \omega t + \phi)
$$

Using the identity:
$$
\sin(\alpha) + \sin(\beta) = 2·\sin[(\alpha + \beta)/2]·\cos[(\alpha − \beta)/2]
$$

We get:
$$
y_{\text{total}}(x,t) = 2A·\cos(\phi/2)·\sin(kx − \omega t + \phi/2)
$$






### 🎯 Interpretation:
- The total wave is a sine wave with **amplitude**:
$$
A_{\text{total}} = 2A·\cos(\phi/2)
$$
- Same frequency and wavelength as originals.

- Same frequency and wavelength as originals.
