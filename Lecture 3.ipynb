## Entanglement

Entanglement is a quantum phenomenon where two or more qubits become connected so that the state of one cannot be described independently of the other. Mathematically speaking - a multi-qubit quantum state is said to be entangled if it cannot be written as a tensor product of individual qubit states.

Let us take a two qubit state, this is separable if it can be written as a tensor product of two states. What is a tensor product- Use I-pad to give some examples.

### Entanglement Recap

A two-qubit state is separable if it can be written as:

$
\ket{\psi} = \ket{\psi_A} \otimes \ket{\psi_B}
$

A two qubit Bell state cannot be written as a tensor prouct of two one qubit state

$
\ket{\Phi_+} = \frac{1}{\sqrt{2}}(\ket{00}+\ket{11})
$

One main property of entangled states is that **Measuring One Qubit Affects the Other**. 

Take the bell state, if we measure qubit A in the computational basis: 
- If the result is 0, the state collapses into $\ket{00}$
- If the result is 1, the state collapses into $\ket{11}$


### Generating entanglement
Entanglement doesnâ€™t happen on its own. Two particles should interact for it to happen. 
- Atoms can become entangled by exchanging photons.
- Electrons can become entangled through spin-spin interactions.
- Qubits in a quantum computer become entangled when engineered circuits cause them to influence each other.

Quantum circuits are generated by gates (similar to classical gates) and coupling the qubits using these gates generate entanglement

#### Single qubit gate
1. Hadamard Gate

This gate creates superposition of the qubits - Recall what is superposition from the first lecture?

$
H \ket{0} = \frac{1}{\sqrt{2}}(\ket{0}+\ket{1})
$

$
H \ket{1} = \frac{1}{\sqrt{2}}(\ket{0}-\ket{1})
$

Recall from lecture 2, every qubit can be expressed as a matrix (row or column depending on if they are bra or ket). Do you remember how to reperesent $\ket{0}$ and $\ket{1}$? In the same way, gates can also be expressed as matrices.

With

$\ket{0} = \begin{pmatrix}
1 \\
0
\end{pmatrix}$ and $\ket{1} = \begin{pmatrix}
0 \\
1
\end{pmatrix}$ 

One may also create the matrix for the single qubit Hadamard gate
$
H = \frac{1}{\sqrt{2}}\begin{pmatrix}
1 & 1\\
1 & -1
\end{pmatrix}
$
